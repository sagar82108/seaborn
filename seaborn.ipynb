```python
#2ans)
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")
print(fmri_data) 
# Step 3: Plot the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.title("Line Plot of Signal over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()
```

```python
#3ans)
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Step 3: Plot the box plots
plt.figure(figsize=(10, 6))

# Box plot for 'age' vs 'pclass'
plt.subplot(1, 2, 1)
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title("Box Plot of Age vs Pclass")
plt.xlabel("Pclass")
plt.ylabel("Age")

# Box plot for 'fare' vs 'pclass'
plt.subplot(1, 2, 2)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title("Box Plot of Fare vs Pclass")
plt.xlabel("Pclass")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```

```python
#4ans)
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Step 3: Plot the histogram with 'cut' as the hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True, bins=30)
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")
plt.legend(title="Cut")
plt.show()
```

```python
#5ans)
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Step 3: Plot the pair plot with 'species' as the hue
sns.pairplot(iris_data, hue="species", markers=["o", "s", "D"])
plt.suptitle("Pair Plot of Iris Dataset by Species", y=1.02)
plt.show()
```

```python
#ans6)
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Step 3: Pivot the dataset to create a matrix form suitable for a heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Step 4: Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", cbar=True)
plt.title("Heatmap of Flight Passengers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
